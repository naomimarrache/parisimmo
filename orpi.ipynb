```python

code_postal=[]
prix=[]
# il faur psliter l'info 
info=[]

nmb_piece=[]

link_appart=[]

div_appart = soup.find_all('li',{'class':'o-grid__col u-flex u-flex-column'})

```

```python
def load_data_by_page(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    for div in div_appart:
        #nb_review.append(div.find('span', {'class':class_nb_review}).text)
        if div.find('div', {'class':class_code_postal}) is not None:
            code_postal.append(div.find('div',attrs={'class':'c-box__inner c-box__inner--sm'}).find('p', attrs={'class':class_code_postal }).find('Paris'))
            prix.append(div.find('strong', attrs={'class':class_prix}))
            info.append(div.find('a',attrs={'class':class_info}))
            link_appart.append(div.find('a', attrs={'class':class_link_appart}).find('/annonce')['href'])
       


    
```

```python

for link in all_appart_links:
    load_data_by_page(link)
    

data = {'code postal':code_postal,'prix':prix,'info':info}


df_orpi = pd.DataFrame(data)
```

```python
def load_data_xpath(link):
    response=requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    code_postal.append(soup.find_element_by_xpath(xpath_code_postal))
    prix.append(soup.find_element_by_xpath(xpath_prix))
    info.append(soup.find_element_by_xpath(xpath_info))            
    link_appart.append(soup.find('a', {'class':class_link_appart}).find('/annonce')['href'])
       
```

```python
load_data_xpath(orpi_link)
    

data = {'code postal':code_postal,'prix':prix,'info':info}


df_orpi = pd.DataFrame(data)
```
