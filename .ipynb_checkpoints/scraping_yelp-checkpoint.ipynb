```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

# Scraping Yelp

```python
#yelp_link = "https://www.yelp.fr/search?find_desc=Restaurants&find_loc=paris&cflt=chinese"
yelp_link = "https://www.yelp.fr/search?find_desc=Restaurants&find_loc=paris&l=p%3AFR-75%3AParis%3A%3ABastille"
class_page_number = "lemon--div__373c0__1mboc border-color--default__373c0__3-ifU text-align--center__373c0__2n2yQ"
response = requests.get(yelp_link)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
number_page = soup.find('div', {"class":class_page_number }).text.split()[2]
```

```python
number_page
```

```python
all_restos_links = []
all_restos_links.append(yelp_link)
```

```python
all_restos_links
```

```python
for i in range(1,int(number_page)):
    all_restos_links.append(yelp_link+"&start="+str(i*30))
```

```python
class_name_resto = "lemon--h4__373c0__1yd__ heading--h4__373c0__27bDo alternate__373c0__2Mge5"
```

```python
name= soup.find('h4', {'class':class_name_resto}).find('a').text
```

```python
name=[]
nb_review=[]
rating=[]
link_resto=[]
```

```python
class_div_restos = "lemon--div__373c0__1mboc container__373c0__3HMKB hoverable__373c0__VqkG7 margin-t3__373c0__1l90z margin-b3__373c0__q1DuY padding-t3__373c0__1gw9E padding-r3__373c0__57InZ padding-b3__373c0__342DA padding-l3__373c0__1scQ0 border--top__373c0__3gXLy border--right__373c0__1n3Iv border--bottom__373c0__3qNtD border--left__373c0__d1B7K border-color--default__373c0__3-ifU" 
class_nb_review = "lemon--span__373c0__3997G text__373c0__2Kxyz reviewCount__373c0__2r4xT text-color--black-extra-light__373c0__2OyzO text-align--left__373c0__2XGa-"
class_rating = "i-stars__373c0__1T6rz"
```

```python
div_restos = soup.find_all('div',{'class':class_div_restos})
```

```python
len(div_restos)
```

```python

def load_data_by_page(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_restos = soup.find_all('div',{'class':class_div_restos})
    for div in div_restos:
        #nb_review.append(div.find('span', {'class':class_nb_review}).text)
        if div.find('span', {'class':class_nb_review}) is not None:
            name.append(div.find('h4', {'class':class_name_resto}).find('a').text)
            nb_review.append(div.find('span', {'class':class_nb_review}).text)
            rating.append(div.find('div',{'class':class_rating})['aria-label'].split()[0])
            link_resto.append("https://www.yelp.fr"+div.find('h4', {'class':class_name_resto}).find('a')['href'])
       
```

```python
for link in all_restos_links:
    load_data_by_page(link)
```

```python
len(link_resto)
```

```python
data = {'name':name,'rating':rating,'nb_reviews':nb_review}
```

```python
df_yelp = pd.DataFrame(data)
```
